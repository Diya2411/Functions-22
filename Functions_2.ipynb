2.  Explain the difference between mutable and immutable data types with examples.
Answer- Mutable Data Types:
Definition: Mutable data types are those whose values can be modified after they are created.
Examples:
->Lists: We can add, remove, or change elements within a list.
->Dictionaries: We can add, remove, or modify key-value pairs in a dictionary.
->Sets: We can add or remove elements from a set.
->User-defined classes: Instances of classes can often be modified.

Immutable Data Types:
Definition:Immutable data types are those whose values cannot be modified after they are created.
Examples:
->Strings: Once a string is created, we cannot change its characters; we can only create a new string.
->Tuples: We cannot change the elements of a tuple once it's created.
Numbers (integers, floats, etc.): Once a number is assigned, its value cannot be changed.
->Booleans: Similar to numbers, booleans are immutable.

3. What are the main differences between lists and tuples in Python.
Answer- Lists:

Syntax: Lists are defined using square brackets [].

Operations: Lists support various methods for modification, such as append(), insert(), remove(), etc.

Syntax: Tuples are defined using parentheses ().

Operations: Tuples have fewer built-in methods compared to lists, as they are not designed for modification.
Memory Usage: Tuples are more memory-efficient than lists because their size is fixed.
Use Cases: Tuples are often used when you need to ensure data integrity and immutability, such as storing coordinates, function return values, or representing fixed sets of data.

11. What advantages do dictionaries offer over lists for certain tasks?
Answer- Dictionaries in Python offer several advantages over lists for certain tasks, particularly when working with key-value pairs. Here are some key advantages:
1. Faster Lookups (O(1) vs. O(n))
Dictionary: Lookup time is O(1) on average because it uses a hash table.
List: Lookup time is O(n) because it requires scanning the entire list to find an element.

2. Meaningful Data Organization
Dictionaries use keys to describe values, making data more readable.
Lists store elements based on position, which may require additional context to understand.

3. Avoiding Duplicate Keys
Dictionaries prevent duplicate keys, ensuring unique identifiers.
Lists allow duplicate values, which may lead to confusion in key-value relationships.

4. Faster Membership Checking (in operator)
Checking if a key exists in a dictionary is O(1), whereas checking in a list is O(n).

5. Data Modification is Easier
Dictionaries allow direct modification using keys, while in lists, you need to find the index first.

14. How does the “in” keyword work differently for lists and dictionaries?
Answer- The in keyword is used to check for the presence of an element in both lists and dictionaries, but it works differently in each case.
1. "in" with a List
When used with a list, in checks whether a value exists in the list.
sandal_list = ["Nike", "Adidas", "Puma"]

print("Reebok" in sandal_list) # False
Searches through all elements in the list.
Time Complexity: O(n) (linear search).

2. "in" with a Dictionary
When used with a dictionary, in checks whether a key exists, not the value.
sandal_dict = {"Nike": 5000, "Adidas": 4500, "Puma": 4000}
print("Nike" in sandal_dict)   # True (checks for key)
print("Reebok" in sandal_dict) # False
print(5000 in sandal_dict)     # False (because it checks only keys)

17. Describe the time complexity of accessing elements in a dictionary.
Answer- Time Complexity of Accessing Elements in a Dictionary (Python)
In Python, dictionaries are implemented using hash tables, which provide average-case constant time complexity for accessing elements.

1. Average-Case Time Complexity: O(1)
When accessing an element using a key (dict[key]), Python computes the hash of the key and looks up the corresponding value.
Since hash table lookups are nearly instantaneous, accessing an element is O(1) on average.

2. Worst-Case Time Complexity: O(n)
In rare cases, if multiple keys have the same hash value (hash collisions), Python resolves them using a linked list or another structure.
If all keys collide, searching for a key may require scanning multiple elements, making it O(n) in the worst case.

Example of Accessing a Dictionary Element
data = {"name": "Amit", "age": 25, "city": "Bangalore"}
print(data["age"])  # O(1) on average

18. In what situations are lists preferred over dictionaries?
Answer- Situations Where Lists Are Preferred Over Dictionaries
While dictionaries offer fast lookups with key-value pairs, lists are preferred in certain situations due to their ordered nature and sequential data storage.

1. Maintaining Order
Use a list when the order of elements matters.
Lists preserve insertion order (since Python 3.7+), whereas dictionaries are unordered collections before Python 3.7.
Example:
fruits = ["Apple", "Banana", "Cherry"]
print(fruits[0])  # Output: Apple (maintains order)

2. Sequential Access & Iteration
If you need to iterate over elements frequently, lists are better.
Accessing elements sequentially in a list (for loop) is efficient.
Example:
for item in [10, 20, 30, 40]:
    print(item)  # Easy iteration

3. Memory Efficiency
Lists are more memory-efficient than dictionaries because they store only values, whereas dictionaries store key-value pairs and require extra space for hashing.
If memory usage is a concern and you don’t need key-value mapping, lists are better.

4. Simpler Data Structures
When you only need a simple collection of elements without labels (keys), a list is sufficient.
For example, storing student roll numbers:
roll_numbers = [101, 102, 103, 104]  # No need for keys

5. Storing Duplicate Values
Lists allow duplicate values, but dictionaries do not.
If you need repeated values, use a list.
Example:
names = ["Alice", "Bob", "Alice", "Charlie"]

When NOT to Use a List (Prefer a Dictionary Instead)
When you need fast lookups (O(1) for dictionaries, O(n) for lists).
When data requires key-value pairs (e.g., mapping student names to grades).
When working with unique keys (dictionaries enforce uniqueness).

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Answer- Why Dictionaries are Unordered:
->Focus on Key-Based Access:
Dictionaries are designed for fast lookups based on keys, not for maintaining a specific sequence of items.

-> Underlying Implementation:
Dictionaries are often implemented using hash tables, which prioritize efficient key-based access over maintaining insertion order.

-> No Index-Based Access:
Unlike lists or tuples, you cannot access elements in a dictionary using an index (e.g., my_dict[2]). Instead, you access them using keys (e.g., my_dict["key"]).

How Unordered Nature Affects Data Retrieval:

-> Order is Not Guaranteed:
When you iterate through a dictionary (e.g., using a for loop), the order in which you get the keys (and therefore the values) is not guaranteed to be the same as the order in which they were inserted.

-> No Position-Based Access:
You cannot rely on the position of a key-value pair to retrieve it, as the position is not a meaningful property of a dictionary.

-> Implications for Certain Operations:
If you need to process dictionary elements in a specific order, you will need to sort the keys or values before iterating or accessing them.

-> Python 3.7+ and Insertion Order:
While dictionaries in Python 3.7 and later versions do preserve insertion order as an implementation detail, it's still considered unordered, and you shouldn't rely on this behavior for all implementations.

List: Always maintains insertion order.

Dictionary: Maintains insertion order from Python 3.7 onward.

4. Storage Structure:

List: Stores only values in a sequence.

Dictionary: Stores key-value pairs, allowing efficient data mapping.

5. Best Use Cases:

List: Ideal for ordered collections, sequential access, and indexing.

Dictionary: Best for fast lookups, mapping relationships, and storing labeled data.
